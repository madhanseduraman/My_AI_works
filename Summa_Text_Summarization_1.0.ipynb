# Extraction based Summarization

### Importing python libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
```

### Reading the Input File

```python
with open(r'C:\Madhan\Analytics\Machine_Learning_project_work\AI_ML\Text_Summarization\interview_summary\interview.txt', "r") as f:
    toy_text = f.read()
```

### Text Cleaning Process

```python
cleaned_text = BeautifulSoup(text_file).get_text() 
letters_only = re.sub("[^a-zA-Z]", " ", cleaned_text) ## Retaining the beginning alphabets and removing everything else
words = letters_only.lower().split()                             
stops = set(stopwords.words("english"))                  
meaningful_words = [w for w in words if not w in stops]   
meaningful_words = ( " ".join( meaningful_words )) 
```

### Word Tokenizing

```python
w_toy_text = nltk.word_tokenize(meaningful_words)
len(w_toy_text)
dist = nltk.FreqDist(w_toy_text)
dist.tabulate
vocab = dist.keys()
```

```python
vocab
```

### LEMMATIZING OF WORDS

```python
WNLemma = nltk.WordNetLemmatizer()
```

```python
w_toy_text_1 = [WNLemma.lemmatize(t) for t in w_toy_text]
len(w_toy_text_1)
dist = nltk.FreqDist(w_toy_text_1)
dist
```

### Count Frequent Words used in the document

```python
freqTable = dict()
for word in w_toy_text_1:
    word = word.lower()
    if word in freqTable:
        freqTable[word] +=1
    else:
        freqTable[word]=1
```

### Score sentences based on Frequent words

```python
sentences = nltk.sent_tokenize(text_file)
sentenceValue = dict()
for sentence in sentences:
    for wordValue in freqTable:
        if wordValue in sentence.lower():
            if sentence[:12] in sentenceValue:
                sentenceValue[sentence[:12]] += wordValue
            else:
                sentenceValue[sentence[:12]] = wordValue
sumValues = 0
for sentence in sentenceValue:
    sumValues += len(sentenceValue[sentence])

# Average value of a sentence from original text
average = int(sumValues/ len(sentenceValue))
```

### Ranking the sentences based on the order & Score

```python
summary = ''
for sentence in sentences:
        if sentence[:12] in sentenceValue and len(sentenceValue[sentence[:12]]) > (1.5 * average):
            print(sentence)
            summary +=  " " + sentence
```

### Writing the output file as .DOC file

```python
with open (r'interview_summary\interview_summary.doc', 'w') as f:
    f.write(summary)
```
